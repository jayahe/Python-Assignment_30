### Problem Statement

In this assignment students have to compress racoon grey scale image into 5 clusters. In
the end, visualize both raw and compressed image and look for quality difference.
The raw image is available in spicy.misc package with the name **face**.

```python
import numpy as np
from sklearn.cluster import  KMeans
import scipy.misc as msc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = msc.face(gray=True)
plt.imshow(img, cmap=plt.cm.gray)
```

```python
img.shape
```

```python
img_r = (img/255.0).reshape(-1,1)
img_r.shape
```

```python
k_colors = KMeans(n_clusters=5).fit(img_r)
y_pred = k_colors.predict(img_r)
```

```python
k_colors.cluster_centers_.shape
```

```python
k_colors.labels_.shape
```

```python
newimg = k_colors.cluster_centers_[k_colors.labels_]
```

```python
newimg.shape
```

```python
newimg = np.reshape(newimg, (img.shape))
```

```python
newimg.shape
```

```python
fig = plt.figure(figsize=(10,8))
ax=fig.add_subplot(1,2,1,xticks=[],yticks=[],title='Original Image')
ax.imshow(img,cmap=plt.cm.gray)
ax=fig.add_subplot(1,2,2,xticks=[],yticks=[],title='Color Compressed Image using K-Means')
ax.imshow(newimg,cmap=plt.cm.gray)
plt.show()
```
